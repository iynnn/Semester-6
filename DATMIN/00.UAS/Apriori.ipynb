```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = [
    'Blouse',
    'Shoes, Skirt, TShirt',
    'Jeans, TShirt',
    'Jeans, Shoes, TShirt',
    'Jeans, Shorts',
    'Shoes, TShirt',
    'Jeans, Skirt',
    'Jeans, Shoes, Shorts, TShirt',
    'Jeans',
    'Jeans, Shoes, TShirt',
    'TShirt',
    'Blouse, Jeans, Shoes, Skirt, TShirt',
    'Jeans, Shoes, Shorts, TShirt',
    'Shoes, Skirt, TShirt',
    'Jeans, TShirt',
    'Skirt, TShirt',
    'Blouse, Jeans, Skirt',
    'Jeans, Shoes, Shorts, TShirt',
    'Jeans',
    'Jeans, Shoes, Shorts, TShirt'
]

data
```

## Olah Data
Ubah data menjadi satu baris per item

```python
lst = list()
for i, items in zip(range(len(data)), data):
    for item in items.split(', '):
        lst.append([i+1, item, 1])

lst = pd.DataFrame(lst).rename(columns = {0 : 'ID', 1 : 'Item', 2 : 'Quantity'})
lst
```

```python
bucket = (lst.groupby(['ID', 'Item'])['Quantity']
          .sum()
          .unstack()
          .reset_index()
          .fillna(0)
          .set_index('ID')
          .applymap(lambda x : 1 if x > 0 else 0))
bucket
```

sebelum data dimasukkan ke dalam algoritma apriori, kita bisa bentuk data seperti tabel di atas

```python

frequent_itemsets = apriori(bucket, min_support=0.30, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)
rules
```
